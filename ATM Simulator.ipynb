```python
# Dummy PIN
correct_pin = 1234

# Function to check if a PIN is valid (4-digit number)
def is_valid_pin(pin):
    return len(str(pin)) == 4 and pin >= 1000 and pin <= 9999

# Main ATM simulator function
def atm_simulator():
    print("***** Welcome To ATM Machine Simulator *****")

    # Initialize pin_attempts to 3 to allow three attempts
    pin_attempts = 3

    while pin_attempts > 0:
        # Prompt the user to enter their PIN
        pin = int(input("Enter Your PIN: "))

        # Check if the entered PIN is a valid 4-digit number
        if is_valid_pin(pin):
            # If it's valid, check if it matches the correct PIN
            if pin == correct_pin:
                # Correct PIN entered, provide access to the menu
                print("PIN is correct. Access granted.")
                balance = 1000

                while True:
                    print("\nOptions you can Exercise are:")
                    print("1) Balance")
                    print("2) Withdraw")
                    print("3) Deposit")
                    print("4) Exit")

                    try:
                        choice = int(input("Select Your Transaction from the above options: "))

                        if choice == 1:
                            print(f'Available A/C Balance: ${balance}')

                        elif choice == 2:
                            withdraw = int(input("Enter Amount to Withdraw: $"))
                            if withdraw <= balance:
                                balance -= withdraw
                                print(f'You have withdrawn: ${withdraw}')
                                print(f'Your available balance is: ${balance}')
                            else:
                                print("Insufficient funds. Transaction canceled.")

                        elif choice == 3:
                            deposit = int(input("Enter Amount to Deposit: $"))
                            if deposit > 0:
                                balance += deposit
                                print(f'You have deposited: ${deposit}')
                                print(f'Your available balance is: ${balance}')
                            else:
                                print("Invalid deposit amount. Transaction canceled.")

                        elif choice == 4:
                            print("Thanks for using ATM Simulator. Have a nice day!")
                            break
                        else:
                            print("Invalid choice. Please select a valid option (1-4).")

                    except ValueError:
                        print("Invalid input. Please enter a valid number.")
                break  # Exit the while loop if the user has completed their transactions

            else:
                # Incorrect PIN entered
                pin_attempts -= 1
                if pin_attempts > 0:
                    print(f"Invalid PIN. You have {pin_attempts} attempts left.")
                else:
                    print("Incorrect PIN. Access denied.")
                    break
        else:
            # Invalid PIN format
            print("Invalid PIN format. Please enter a valid 4-digit PIN.")
       
if __name__ == "__main__":
    atm_simulator()

```

# ATM Simulator Documentation

Logic:
This Python script simulates an ATM machine. It prompts the user to enter a 4-digit PIN.
Upon correct entry, it provides access to account transactions like balance inquiry,
withdrawal, and deposit. The user can also choose to exit the simulator.

Functions:
- is_valid_pin(pin): Checks if a PIN is a valid 4-digit number.
- atm_simulator(): Main function that runs the ATM simulation.

Real-time Usage:
This code can be used as a basic ATM simulator for educational purposes. It demonstrates
PIN validation, user interaction, and transaction handling.
